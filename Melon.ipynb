```python
# 크롬 브라우저의 데이터를 읽어오는 방법

from selenium import webdriver

driver = webdriver.Chrome('chromedriver.exe')  # 파이썬으로 브라우저 컨트롤
url = "https://www.melon.com/chart/index.htm"
driver.get(url)
html = driver.page_source
soup = bs(html)
```

```python
song_data = []
rank = 1

song = songs = soup.select('tr.lst50')

for song in songs:
    title = song.select('span > a')[0].text
    singer = song.select('span > a')[1].text
    song_data.append(['Melon', rank, title, singer])
    rank += 1
    
song_data

song = songs = soup.select('tr.lst100')

for song in songs:
    title = song.select('span > a')[0].text
    singer = song.select('span > a')[1].text
    song_data.append(['Melon', rank, title, singer])
    rank += 1
    
song_data
```
