```python
import requests # 웹 연결 관련 라이브러리
```

```python
from bs4 import BeautifulSoup as bs # 분석을 용이하게 다듬어주는 라이브러리
```

```python
import pandas as pd  # 데이터 분석 관련 라이브러리
```

```python
url = "https://music.bugs.co.kr/chart" # 벅스, 지니, 멜론
html = requests.get(url)
html.text
```

```python
soup = bs(html.text,'html.parser')
```

```python
print(soup)
```

```python
len(soup.select('tr')) # html tag(code)를 사용할때는 select 사용
```

```python
songs = soup.select('tr')
len(songs)
```

```python
print(songs)
```

```python
songs = soup.select('tbody>tr')
len(songs)                    
```

```python
song_data = []
rank = 1

songs = soup.select('table.byChart> tbody >tr')

for song in songs:
    title = song.select('p.title > a')[0].text
    singer = song.select('p.artist > a')[0].text
    song_data.append(['Bugs', rank, title, singer])
    rank += 1 # rank = rank + 1

song_data
```

```python
df = pd.DataFrame(song_data)
df
```

```python
df.to_excel('bugs.xlsx', index=False)
```
